You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column

````
np.random.seed(0)
X = np.random.rand(10,5)
print(X)
print(np.amin(X,axis=0)) #Minimum value of each column
print(np.argmin(X,axis=0)) #Index of minimum value of each column
print(np.amax(X,axis=0)) #Maximum value of each column
print(np.argmax(X,axis=0)) #Index of maximum value of each column
print(np.amin(X,axis=1)) #Minimum value of each row
print(np.argmin(X,axis=1)) #Index of minimum value of each row
print(np.amax(X,axis=1)) #Maximum value of each row
print(np.argmax(X,axis=1)) #Index of maximum value of each row
print(np.amin(X)) #Minimum value of entire matrix
print(np.argmin(X)) #Minimum value of entire matrix
import math
print("Row=",math.floor(np.argmin(X)/5),"col=",math.floor(np.argmin(X)%5)) #row column index of minimum value of entire matrix
print(np.amax(X)) #Minimum value of entire matrix
print(np.argmax(X)) #Minimum value of entire matrix
import math
print("Row=",math.floor(np.argmax(X)/5),"col=",math.floor(np.argmax(X)%5)) #row column index of minimum value of entire matrix
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
import pandas as pd
train = pd.read_csv("titanic3.csv")
#train['sex'] = train.sex.map({'female':0, 'male':1})
train.head()
````

````
train['embarked'].value_counts()
````

````
sex_dummies = pd.get_dummies(train.sex, prefix='sex').iloc[:, :]
train = pd.concat([train,sex_dummies], axis=1)
````

````
embarked_dummies = pd.get_dummies(train.embarked, prefix='embarked').iloc[:, :]
train = pd.concat([train, embarked_dummies], axis=1)
````

````
pclass_dummies = pd.get_dummies(train.pclass, prefix='pclass').iloc[:, :]
train = pd.concat([train, pclass_dummies], axis=1)
````

````
train.head()
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
train['age'] = train['age'].fillna(train['age'].mean())
train['fare'] = train['fare'].fillna(train['fare'].mean())
````

````
train.head()
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
from sklearn import preprocessing
# Create x, where x the 'scores' column's values as floats
x = train[['age']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
df_normalized = pd.DataFrame(x_scaled)
df = pd.DataFrame({'NormalizedAge':[x[1][0] for x in df_normalized.iterrows()]})

#print(df_normalized)
train = pd.concat([train, df], axis=1)
````

````
for x in df_normalized.iterrows():
  print(x[1][0])
````

````
train.head()
````

````
train = train.drop('NormalizedAge',axis=1)
````

````
from sklearn import preprocessing
# Create x, where x the 'scores' column's values as floats
x = train[['fare']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
df_normalized = pd.DataFrame(x_scaled)
df = pd.DataFrame({'NormalizedFare':[x[1][0] for x in df_normalized.iterrows()]})

#print(df_normalized)
train = pd.concat([train, df], axis=1)
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
# Get x values of the sine wave
time = np.arange(0, 3*np.pi, 0.1)
# Amplitude of the sine wave is sine of a variable like time
amplitude = np.sin(time)
# Plot a sine wave using time and amplitude obtained for the sine wave
plot.plot(time, amplitude)
# Give a title for the sine wave plot
plot.title(label=r'$\sin (x)$')
# Give x axis label for the sine wave plot
plot.xlabel('Time')
# Give y axis label for the sine wave plot
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
fig= plot.figure(figsize=(10,10))
plot.show()
# Display the sine wave
plot.show()
````
